# Visualization 4
a scatter plot of classification vs each category

```python
plt.figure(figsize=(12,12)) 

plt.subplot(4,1,1)
plt.scatter(iris['Classification'],iris['SL'],s=10,c='green')
plt.title('Plot of Classification vs SL',fontsize=10)
plt.xlabel('Classification',fontsize=10)
plt.ylabel('SL',fontsize=10)

plt.subplot(4,1,2) 
plt.scatter(iris['Classification'],iris['SW'],s=10,c='BLUE')
plt.title('Plot of Classification vs SW',fontsize=10)
plt.xlabel('Classification',fontsize=10)
plt.ylabel('SW',fontsize=10)

plt.subplot(4,1,3) 
plt.scatter(iris['Classification'],iris['PL'],s=10,c='red')
plt.title('Plot of Classification vs PL',fontsize=10)
plt.xlabel('Classification',fontsize=10)
plt.ylabel('PL',fontsize=10)

plt.subplot(4,1,4) 
plt.scatter(iris['Classification'],iris['PW'],s=10,c='orange')
plt.title('Plot of Classification vs PW',fontsize=10)
plt.xlabel('Classification',fontsize=10)
plt.ylabel('PW',fontsize=10)

plt.tight_layout()
```
